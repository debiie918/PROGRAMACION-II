```python
#cambiar el flujo normal de un cilclo con el uso de continue
frutas_1= ["banana","manzana","kiwi","pera","ciruela","uva"]

for frutas in  frutas_1:
    if  frutas == "ciruela" or frutas =="pera":
        continue
    print(f"voy a comer una {frutas}")

```

```python
#ejemplo con el break
for frutas in frutas_1:
    if frutas == "ciruela":
        break
    print(f"voy a comer una {frutas}")
```
